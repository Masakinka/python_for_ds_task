### Завдання 1: Z-тест для однієї вибірки

**Мета**: Перевірити, чи середнє значення вибірки відрізняється від гіпотетичного середнього.

**Припущення**: Середній ріст студентів університету вважається 170 см. Ми хочемо перевірити, чи середній ріст випадкової вибірки студентів істотно більший. Вибірка задана в змінній `heights`.

Для виконання завдання імпортуйте метод для проведення `ztest`, виконайте Z-тест з Python, виведіть p-value і зробіть заключення, чи справдилось припущення про середній зріст студентів на рівні значущості 0.05?

Використовуємо тут Z-тест, оскільки вибірка достатньо велика (більше 30 екземплярів) і можемо вважати, що її стандартне відхилення близьке до ст. відхилення генеральної сукупності, а значить - відоме.

```python
import numpy as np
from statsmodels.stats.weightstats import ztest
from scipy.stats import ttest_ind, norm
```

```python
heights = [174, 171, 175, 179, 170, 170, 179, 175, 169, 174, 169, 169, 173, 162, 163, 169, 166,
           173, 167, 164, 179, 170, 172, 164, 169, 175, 169, 174, 169, 169, 173, 162, 177]
```

```python
np.mean(heights)
```

```python
ztest_Score, p_value = ztest(heights, value = 170, alternative='two-sided')
```

```python
ztest_Score, p_value
```

```python
if p_value < 0.05:
    print("Висновок: середній ріст студентів істотно більший за 170 см з рівнем значущості 0.05.")
else:
    print("Висновок: немає статистично значущих доказів, що середній ріст студентів більший за 170 см з рівнем значущості 0.05.")
```

1. Тестова статистика негативна, що вказує на те, що середнє значення у вибірці менше, ніж гіпотетичні 170 см, хоча завдання полягає в тому, щоб перевірити, чи більше середнє зроста. Встановлення альтернативної гіпотези ('two-sided') відповідає негативної тестової статистиці.
2.  Враховуючи завдання та напрямок тесту (односторонній, перевірка на те, чи більше середнє), p-значення близьке до 1.0 (округлене у виведенні 0.8482235371587619) фактично підтримує нульову гіпотезу. Це вказує на те, що середный зрост не значно більший за 170 см.

Оскільки p-value < 0.05, у нас є достатньо доказів, то значення p-value свідчить про те, що середні оцінки між двома групами статистично значимо різняться. Це дозволяє відкинути нульову гіпотезу у тому, що відмінностей немає, і підтвердити, що новий курс ефективніший проти стандартним.

### Завдання 3\*: Z-тест для двох **пов'язаних** вибірок

**Мета**: Перевірити, чи вплинув тренінг на продуктивність працівників, коли всі справробітники проходили тренінг.

**Припущення**: Наші працівники проходять однаковий тренінг, і ми хочемо з'ясувати, чи покращилися їхні показники після тренінгу на рівні значущості 0.05. Виконайте Z-тест з Python, виведіть p-value і зробіть заключення, чи поліпшились показники працівників на рівні значущості 0.05?

Зверніть увагу, ці вибірки повʼязані між собою, значить не буде коректно виконати z-test між ними, натомість, ми можемо порівняти, чи різниця в результатах (продуктивності співробітників) є стат значущою.

```python
before_training = np.array([57.82, 37.63, 36.8 , 55.22, 52.97, 52.5 , 53.46, 43.2 , 52.32,
       52.93, 42.86, 68.66, 54.74, 38.09, 56.57, 40.25, 57.87, 61.59,
       41.79, 59.63, 54.13, 58.22, 68.97, 47.55, 42.46, 41.1 , 41.84,
       49.23, 53.41, 52.77])
after_training = np.array([62.47, 40.66, 42.7 , 57.69, 61.41, 56.76, 54.75, 44.06, 56.29,
       55.48, 47.28, 72.6 , 57.59, 39.39, 56.54, 42.36, 62.58, 65.01,
       42.3 , 62.98, 57.9 , 59.45, 72.28, 50.66, 43.18, 44.82, 45.96,
       54.4 , 58.52, 53.01])
```

Для аналізу впливу тренінгу на продуктивність працівників за допомогою z-тесту для пов'язаних вибірок, потрібно спочатку визначити різницю між двома наборами даних (до і після тренінгу), а потім перевірити, чи є ця різниця статистично значущою.

```python
# Розрахунок різниці
differences = after_training - before_training
differences
```

```python
# Розрахунок середнього і стандартного відхилення різниці
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)  # ddof=1 для незміщеної оцінки стандартного відхилення
mean_diff, std_diff
```

```python
# Кількість спостережень
n = len(differences)
n
```

```python
# Z-тест
z_score = mean_diff / (std_diff / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(np.abs(z_score)))  # Two-tailed test
z_score, p_value
```

Z-score дорівнює 9.1389, що є високим значенням. Це означає, що середнє покращення показників продуктивності віддалене від нуля на 9.1389 стандартних відхилень. Таке значне відхилення вказує на те, що тренінг мав значний вплив на покращення показників.

P-value дорівнює 0.0, що вказує на вкрай малу ймовірність отримати таку або більш значущу різницю між двома вибірками, якщо б припущення про відсутність впливу тренінгу було б вірним. Значення p-value менше порога значущості 0.05 дає нам підстави відкинути нульову гіпотезу.

Висновок:
Тренінг мав статистично значущий позитивний вплив на продуктивність працівників. Результати показують, що після проходження тренінгу показники продуктивності працівників значно покращились. Ці результати можуть бути використані для аргументації користі від тренінгу та його подальшого впровадження серед інших груп співробітників або для подальшого розширення цієї програми тренінгу.
