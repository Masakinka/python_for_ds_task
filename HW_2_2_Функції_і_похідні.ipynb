### –î–æ–º–∞—à–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ –≤—ñ–¥–ø—Ä–∞—Ü—é–≤–∞–Ω–Ω—è –ø–æ–Ω—è—Ç—Ç—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó

1. –ó–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –æ—Ü—ñ–Ω–∫–∏ –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω, –≤–∏—Ç—Ä–∞—á–µ–Ω–∏—Ö –Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è: $ f(x) = 10x - 5 $.
  1. –ó –¥–æ–ø–æ–º–æ–≥–æ—é Python –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó.
  2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É `matplotlib` –¥–ª—è –ø–æ–±—É–¥–æ–≤–∏ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –∑–Ω–∞—á–µ–Ω—å –≤—ñ–¥ 0 –¥–æ 10 –≥–æ–¥–∏–Ω. –ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ, —è–∫ –∑–º—ñ–Ω–∞ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω –≤–∏–≤—á–µ–Ω–Ω—è –≤–ø–ª–∏–≤–∞—î –Ω–∞ –æ—Ü—ñ–Ω–∫—É.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# –í–∏–∑–Ω–∞—á–∞—î–º–æ —Å–∏–º–≤–æ–ª—ñ—á–Ω—É –∑–º—ñ–Ω–Ω—É
x_sym = sp.symbols('x')

# –õ—ñ–Ω—ñ–π–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è
f_lin = 10 * x_sym - 5
derivative_f_lin = sp.diff(f_lin, x_sym)

print("–ü–æ—Ö—ñ–¥–Ω–∞ –ª—ñ–Ω—ñ–π–Ω–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó 10 * x -5:", derivative_f_lin)
```

```python
# –ü–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ —Ñ—É–Ω–∫—Ü—ñ—é —É –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—é –¥–ª—è –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –∑ numpy
f_lin_np = sp.lambdify(x_sym, f_lin, 'numpy')

# –í–∏–∑–Ω–∞—á–µ–Ω–Ω—è –¥—ñ–∞–ø–∞–∑–æ–Ω—É –∑–Ω–∞—á–µ–Ω—å –¥–ª—è x
x = np.linspace(0, 10, 400)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(10, 5))
plt.plot(x, f_lin_np(x), label='f(x) = 10 * x - 5', color='blue')
plt.axhline(y=10, color='green', linestyle='-', label="f'(x) = 10")

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –ª–µ–≥–µ–Ω–¥–∏, –º—ñ—Ç–æ–∫ —Ç–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

–ì—Ä–∞—Ñ—ñ–∫ –ø–æ–∫–∞–∑—É—î, —â–æ —Ñ—É–Ω–∫—Ü—ñ—è
ùëì(ùë•)=10ùë•‚àí5 —î –ª—ñ–Ω—ñ–π–Ω–æ—é, —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–∞ ùëì‚Ä≤(ùë•)=10 —î –∫–æ–Ω—Å—Ç–∞–Ω—Ç–æ—é. –¶–µ –æ–∑–Ω–∞—á–∞—î, —â–æ –∑–±—ñ–ª—å—à–µ–Ω–Ω—è –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω –Ω–∞–≤—á–∞–Ω–Ω—è –Ω–∞ 1 –≥–æ–¥–∏–Ω—É –∑–±—ñ–ª—å—à—É—î –æ—Ü—ñ–Ω–∫—É –Ω–∞ 10 –±–∞–ª—ñ–≤.

2. –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ Python —ñ `SymPy`, –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—ñ –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ–π $ f(x) = \sin(x) $ —Ç–∞ $ f(x) = \ln(x) $. –ó–æ–±—Ä–∞–∑—ñ—Ç—å —Ü—ñ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—Ö–Ω—ñ –ø–æ—Ö—ñ–¥–Ω—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é `matplotlib`.



```python
# –§—É–Ω–∫—Ü—ñ—è f(x) = sin(x)
f_sin = sp.sin(x_sym)
derivative_f_sin = sp.diff(f_sin, x_sym)

# –§—É–Ω–∫—Ü—ñ—è f(x) = ln(x)
f_ln = sp.ln(x_sym)
derivative_f_ln = sp.diff(f_ln, x_sym)

print("–ü–æ—Ö—ñ–¥–Ω–∞ –ª—ñ–Ω—ñ–π–Ω–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó sin(x):", derivative_f_sin)
print("–ü–æ—Ö—ñ–¥–Ω–∞ –ª—ñ–Ω—ñ–π–Ω–æ—ó —Ñ—É–Ω–∫—Ü—ñ—ó ln(x):", derivative_f_ln)
```

```python
# –ü–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—Ö–Ω—ñ –ø–æ—Ö—ñ–¥–Ω—ñ —É –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—ó –¥–ª—è –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –∑ numpy
f_sin_np = sp.lambdify(x_sym, f_sin, 'numpy')
derivative_f_sin_np = sp.lambdify(x_sym, derivative_f_sin, 'numpy')

f_ln_np = sp.lambdify(x_sym, f_ln, 'numpy')
derivative_f_ln_np = sp.lambdify(x_sym, derivative_f_ln, 'numpy')

# –í–∏–∑–Ω–∞—á–µ–Ω–Ω—è –¥—ñ–∞–ø–∞–∑–æ–Ω—É –∑–Ω–∞—á–µ–Ω—å –¥–ª—è x
x_sin = np.linspace(-2 * np.pi, 2 * np.pi, 400)
x_ln = np.linspace(0.1, 10, 400)
```

```python
# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É –¥–ª—è sin(x) —Ç–∞ –π–æ–≥–æ –ø–æ—Ö—ñ–¥–Ω–æ—ó
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(x_sin, f_sin_np(x_sin), label='f(x) = sin(x)', color='blue')
plt.plot(x_sin, derivative_f_sin_np(x_sin), label="f'(x) = cos(x)", color='green')
plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó sin(x) —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É –¥–ª—è ln(x) —Ç–∞ –π–æ–≥–æ –ø–æ—Ö—ñ–¥–Ω–æ—ó
plt.subplot(1, 2, 2)
plt.plot(x_ln, f_ln_np(x_ln), label='f(x) = ln(x)', color='blue')
plt.plot(x_ln, derivative_f_ln_np(x_ln), label="f'(x) = 1/x", color='green')
plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó ln(x) —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```


3. –ó–∞–≥–∞–ª—å–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏ –∫–æ–º–ø–∞–Ω—ñ—ó –≤—ñ–¥ –æ–±—Å—è–≥—É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –∑–∞–¥–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—î—é $ f(x) = x^2 - 5x + 1 $, –¥–µ $ x $ ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å. –û–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –¥–æ–ø–æ–º–æ–≥–æ—é SymPy —Ç–∞ –≤—ñ–∑—É–∞–ª—ñ–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –Ω–∞ –≤—ñ–¥—Ä—ñ–∑–∫—É –≤—ñ–¥ 0 –¥–æ 100.

```python
# –§—É–Ω–∫—Ü—ñ—è f(x) = x^2 - 5x + 1
f = x_sym**2 - 5*x_sym + 1
derivative_f = sp.diff(f, x_sym)

print("–ü–æ—Ö—ñ–¥–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—ó x^2 - 5x + 1:", derivative_f)
```

```python
# –ü–µ—Ä–µ—Ç–≤–æ—Ä—é—î–º–æ —Ñ—É–Ω–∫—Ü—ñ—é —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω—É —É –ª—è–º–±–¥–∞-—Ñ—É–Ω–∫—Ü—ñ—ó –¥–ª—è –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –∑ numpy
f_np = sp.lambdify(x_sym, f, 'numpy')
derivative_f_np = sp.lambdify(x_sym, derivative_f, 'numpy')

# –í–∏–∑–Ω–∞—á–µ–Ω–Ω—è –¥—ñ–∞–ø–∞–∑–æ–Ω—É –∑–Ω–∞—á–µ–Ω—å –¥–ª—è x
x = np.linspace(0, 100, 400)
```

```python
# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(10, 5))
plt.plot(x, f_np(x), label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x, derivative_f_np(x), label="f'(x) = 2x - 5", color='green')

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –ª–µ–≥–µ–Ω–¥–∏, –º—ñ—Ç–æ–∫ —Ç–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

–ì—Ä–∞—Ñ—ñ–∫ –ø–æ–∫–∞–∑—É—î, —è–∫ —Ñ—É–Ω–∫—Ü—ñ—è x^2 - 5x + 1 –∑–º—ñ–Ω—é—î—Ç—å—Å—è –∑ –æ–±—Å—è–≥–æ–º –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ (–∫—ñ–ª—å–∫—ñ—Å—Ç—é –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å) —ñ —è–∫ —ó—ó –ø–æ—Ö—ñ–¥–Ω–∞ f ‚Ä≤(x)=2x‚àí5 –≤–∏–∑–Ω–∞—á–∞—î —à–≤–∏–¥–∫—ñ—Å—Ç—å –∑–º—ñ–Ω–∏ –∑–∞–≥–∞–ª—å–Ω–∏—Ö –≤–∏—Ç—Ä–∞—Ç –≤ –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å.
