1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Визначаємо символічну змінну
x_sym = sp.symbols('x')

# Лінійна функція
f_lin = 10 * x_sym - 5
derivative_f_lin = sp.diff(f_lin, x_sym)

print("Похідна лінійної функції 10 * x -5:", derivative_f_lin)
```

```python
# Перетворюємо функцію у лямбда-функцію для використання з numpy
f_lin_np = sp.lambdify(x_sym, f_lin, 'numpy')

# Визначення діапазону значень для x
x = np.linspace(0, 10, 400)

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f_lin_np(x), label='f(x) = 10 * x - 5', color='blue')
plt.axhline(y=10, color='green', linestyle='-', label="f'(x) = 10")

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Графік показує, що функція
𝑓(𝑥)=10𝑥−5 є лінійною, і її похідна 𝑓′(𝑥)=10 є константою. Це означає, що збільшення кількості годин навчання на 1 годину збільшує оцінку на 10 балів.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# Функція f(x) = sin(x)
f_sin = sp.sin(x_sym)
derivative_f_sin = sp.diff(f_sin, x_sym)

# Функція f(x) = ln(x)
f_ln = sp.ln(x_sym)
derivative_f_ln = sp.diff(f_ln, x_sym)

print("Похідна лінійної функції sin(x):", derivative_f_sin)
print("Похідна лінійної функції ln(x):", derivative_f_ln)
```

```python
# Перетворюємо функції та їхні похідні у лямбда-функції для використання з numpy
f_sin_np = sp.lambdify(x_sym, f_sin, 'numpy')
derivative_f_sin_np = sp.lambdify(x_sym, derivative_f_sin, 'numpy')

f_ln_np = sp.lambdify(x_sym, f_ln, 'numpy')
derivative_f_ln_np = sp.lambdify(x_sym, derivative_f_ln, 'numpy')

# Визначення діапазону значень для x
x_sin = np.linspace(-2 * np.pi, 2 * np.pi, 400)
x_ln = np.linspace(0.1, 10, 400)
```

```python
# Створення графіку для sin(x) та його похідної
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(x_sin, f_sin_np(x_sin), label='f(x) = sin(x)', color='blue')
plt.plot(x_sin, derivative_f_sin_np(x_sin), label="f'(x) = cos(x)", color='green')
plt.title("Графік функції sin(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Створення графіку для ln(x) та його похідної
plt.subplot(1, 2, 2)
plt.plot(x_ln, f_ln_np(x_ln), label='f(x) = ln(x)', color='blue')
plt.plot(x_ln, derivative_f_ln_np(x_ln), label="f'(x) = 1/x", color='green')
plt.title("Графік функції ln(x) та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Функція f(x) = x^2 - 5x + 1
f = x_sym**2 - 5*x_sym + 1
derivative_f = sp.diff(f, x_sym)

print("Похідна функції x^2 - 5x + 1:", derivative_f)
```

```python
# Перетворюємо функцію та її похідну у лямбда-функції для використання з numpy
f_np = sp.lambdify(x_sym, f, 'numpy')
derivative_f_np = sp.lambdify(x_sym, derivative_f, 'numpy')

# Визначення діапазону значень для x
x = np.linspace(0, 100, 400)
```

```python
# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f_np(x), label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x, derivative_f_np(x), label="f'(x) = 2x - 5", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Графік показує, як функція x^2 - 5x + 1 змінюється з обсягом виробництва (кількістю вироблених одиниць) і як її похідна f ′(x)=2x−5 визначає швидкість зміни загальних витрат в залежності від кількості вироблених одиниць.
